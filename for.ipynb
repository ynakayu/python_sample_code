```python
# リストでforを扱おう
# 例題)[10, 20, 30, 40, 50]の
# 各要素を1つずつ出力する。for利用。
scores = [10, 20, 30, 40, 50]
for score in scores:
    print(score)
```

```python
# forの入れ子を使おう
# 例題)九九を出力。forを2つ使う。
for i in range(1, 10):
    for j in range(1, 10):
        print(f'{i} * {j} = {i * j}')
```

```python
# break, continue

# break・・・繰り返し自体を終了する
# continue・・・現在の回だけを中断して、
# 次の回を実行する。

# 例題）九九のプログラムで1の段だけ出力
# for i in range(1, 10):
#     if i == 2:
#         print('break!')
#         break
#     for j in range(1, 10):
#         print(f'{i} * {j} = {i * j}')

# 例題)続き。jが奇数の場合だけ
# 計算結果を出力する
for i in range(1, 10):
    if i == 2:
        print('break!')
        break
    for j in range(1, 10):
        if j % 2 == 0:
            print('continue!')
            continue
        print(f'{i} * {j} = {i * j}')
```

```python
# whileの無限ループ
scores = [10, 20, 30, 40, 50]
count = 0
while count < len(scores):
    print(scores[count])
```

```python
# 例題）
# 1が入力されたら、 Run menu 1.と表示
# 2が入力されたら、  Run menu 2.と表示
# 3が入力されたら、 Run menu 3.と表示
# 9が入力されたら、終了
# 上記以外の数字が入力されたら、
# Invalid numberと出力。再入力を求める。
while True:
    num = int(input('Please enter menu number.(1, 2, 3. Exit: 9)'))
    if num == 1:
        print('Run menu 1.')
    elif num == 2:
        print('Run menu 2.')
    elif num == 3:
        print('Run menu 3.')
    elif num == 9:
        print('Exit!')
        break
    else:
        print('Invalid number!')
```

```python
# 演習: 繰り返し処理を使って、テストの手数の合計値、平均値を求めよう
scores = [100, 90, 80, 70, 60]
sum = 0
for i in scores:
    sum += i

average = sum / len(scores)

print(f'合計値: {sum}')
print(f'平均値: {average}')
```
